prob 1.1: Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
#to write own myreduce function
def myreduce(fun1,lst):
    x=lst[0]
    for i in range(1,len(lst)):
        x=fun1(x,lst[i])
    return x  

#To write a summation function
def sum1(x1,x2):
    return x1+x2 

#To write a product function
def product1(x1,x2):
    return x1*x2
```

```python
#print the sum and product using myreduce function
lst2=[1,2,3,4,5]
print("sum is",myreduce(sum1,lst2))
print("product is",myreduce(product1,lst2))
```
